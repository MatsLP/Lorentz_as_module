```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import lorentz

pars= {'sigma' : 10.,
       'R' : 28.,
       'b' : 8./3.,
       'dt': 0.01,
       'numsteps' : 10000}

outarray = np.zeros((pars['numsteps']+ 1, 4))
outarray[0,:] = np.array([10.,1.,1.,0.])


lorentz.solve(outarray, pars)

plt.plot(outarray[:,0],outarray[:,2])
plt.show()
```
